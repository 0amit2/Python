```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
path=r"C:\Users\amitk\Downloads\TCA_Internship\TCA_Machine_learning\student_info.csv"
df=pd.read_csv(path)
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
median=df.study_hours.median()
```

```python
df['study_hours']=df.study_hours.fillna(median)
```

```python
df.isnull().sum()
```

```python
df
```

```python
plt.scatter(df['study_hours'],df['student_marks'])
plt.xlabel('Study_hours')
plt.ylabel('Student_marks')
plt.show()
```

```python
X=df[['study_hours']]
X
```

```python
y=df['student_marks']
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
predict=model.predict(X_test)
predict[:5]
```

```python
y_test.head()
```

```python
plt.scatter(df['study_hours'],df['student_marks'])
plt.plot(df['study_hours'],model.predict(df[['study_hours']]),color='r',label='regression')
plt.xlabel('Study_hours')
plt.ylabel('Student_marks')
plt.legend()
plt.show()
```
